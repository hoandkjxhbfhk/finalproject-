```python
from facebook_scraper import *
import pandas as pd
import time
import numpy as np
from itertools import groupby
from time import sleep
```

```python
import pickle
with open('all_commenters_ids.pkl', 'rb') as file:
    all_commenters_id = pickle.load(file)

print("Danh sách đã được đọc từ file:", len(all_commenters_id))
```

```python
unique_commenters_id = [key for key, group in groupby(all_commenters_id)]
```

```python
index_of_target_element = unique_commenters_id.index("100051578146240")

# Lấy các phần tử nằm sau phần tử có giá trị "100042032808485"
elements_after_target = unique_commenters_id[index_of_target_element + 1:]

print("Các phần tử sau '100052560993081':", elements_after_target)
```

```python

len(elements_after_target)
```

```python
max, index = 0, 0
user_profiles = []
cookie_path = 'cookies.txt'
for i in range(len(elements_after_target)):
    try:
        profile = get_profile(elements_after_target[i])
        
        # Check if the profile is not empty
        if profile:
            keys_count = len(profile.keys())
            
            # Update max_keys and index if needed
            if len(profile.keys()) > max:
              max = len(profile.keys())
              index = i
            
            user_profiles.append(profile)
            print(profile)
        
    except Exception as e:
        print(f"Error processing user {elements_after_target[i]}: {e}")
        continue

      
```

```python
with open('all_commenters_ids.pkl', 'wb') as file:
    pickle.dump(elements_after_target, file)

print("Danh sách đã được lưu vào elements_after_target.pkl")
```

```python
#Chuyển danh sách thành DataFrame
df = pd.DataFrame(user_profiles)

# Lưu DataFrame vào file CSV
df.to_csv('user_profiles6.csv', index=False)
```
